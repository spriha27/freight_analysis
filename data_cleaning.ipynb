```python
import pandas as pd

df = pd.read_csv("FAF5 Total Truck Flows by Commodity_2022.csv")
restart_df = df
```

```python
df = restart_df
print(df.columns)
```

```python
freight_types = {
    'Farm Products': 'farm',
    'Food, Bev and Tobacco': 'foodandbev',
    'Stone Sand Gravel & Ores': 'construction',
    'Liquid and Gases': 'fluids',
    'Chemicals': 'chemicals',
    'Logs & Oth Wood Prods': 'wood',
    'Waste and Scrap': 'waste',
    'Consumer Manuf': 'consumer_manuf',
    'Durable Manuf (low tech)': 'dur_manuf_low',
    'Durable Manuf (high tech)': 'dur_manuf_high',
    'Motor and Other Veh': 'motorvehicles',
    'Mixed Freight': 'mixed'
}
```

```python
def rename_columns(df):
    column_mapping = {}
    
    for col in df.columns:
        if col == 'ID':
            column_mapping[col] = 'route_id'
            continue
            
        if col.startswith(('AB ', 'BA ', 'TOT ')):
            direction, rest = col.split(' ', 1)
            direction = direction.lower()
            
            if rest in ['Tons_22 All', 'Trips_22 All']:
                measure = rest.split('_')[0].lower()
                column_mapping[col] = f"{direction}_{measure}"
                continue
                
            for ft, short in freight_types.items():
                if ft and f"{ft}-" in rest:
                    measure = rest.split('-')[1].split('_')[0].lower()
                    column_mapping[col] = f"{direction}_{short}_{measure}"
                    break
    
    return df.rename(columns=column_mapping)
```

```python
df_renamed = rename_columns(df)
df_renamed.head()
```

```python
smaller_df = df_renamed.sample(n=10000)
smaller_df.head()
```

```python
smaller_df = smaller_df.dropna()
smaller_df = smaller_df.drop_duplicates()
df_cleaned = smaller_df.reset_index(drop=True)
df_cleaned.to_csv('cleaned_dataset.csv', index=False)
```

```python
reverse_freight_types = {v: k for k, v in freight_types.items() if k}
direction = ['ab', 'ba', 'tot']
def normalize_df(df):
    records = []
    for idx, row in df.iterrows():
        for freight_code, freight_name in reverse_freight_types.items():
                record = {
                    'route_id': row['route_id'],
                    'direction': direction.upper(),
                    'freight_type': freight_name,
                    'tons': row.get(f'{direction}_{freight_code}_tons', None),
                    'trips': row.get(f'{direction}_{freight_code}_trips', None),
                    'total_tons': row.get(f'tot_{freight_code}_tons', None),
                    'total_trips': row.get(f'tot_{freight_code}_trips', None)
                }
                records.append(record) 
```
